# Comparison of several median filter implementations

- `scipy.signal.medfilt2d`
- `scipy.ndimage.median_filter`
- `skimage.filters.median`

```python
import numpy as np
# Pillow
from PIL import Image
import scipy.signal
import scipy.ndimage
# scikit-image
import skimage.filters
from skimage.morphology import square
```

```python
image_rgb = np.asarray(Image.open('../data/source_images/IMG_3262_denoised.jpg'), dtype='uint8')
image = image_rgb.mean(axis=-1).astype('uint8')
```

```python
image.dtype
```

```python
Image.fromarray(image)
```

```python
kernel_size = 31
```

```python
%time image_med_scipy_signal = scipy.signal.medfilt2d(image, kernel_size=kernel_size)
Image.fromarray(image_med_scipy_signal)
```

```python
%time image_med_scipy_ndimage = scipy.ndimage.median_filter(image, size=kernel_size)
Image.fromarray(image_med_scipy_ndimage)
```

```python
%time image_med_skimage = skimage.filters.median(image, selem=square(kernel_size))
Image.fromarray(image_med_skimage)
```

```python
%time image_med_skimage = skimage.filters.median(image, selem=square(100))
Image.fromarray(image_med_skimage)
```
